```python
@njit
def score_board(board):
    # for each row, column and diagonal, we add points based on how many 1s/-1s there are in a row
    # specifically, we split up into all possible 3 in a rows and go through them one by one
    # if all 3 are the same, we give 9 points*player
    # if 2 are the same and 1 is empty, we give 4 point*player
    # if 2 are the same and 1 is the opponent, we give 1 point*player
    # if 1 is the same and 2 are empty, we give 1 point*player
    # note! If a whole row is filled with 1s/-1s, we give 0 points
    # TODO: make this better, it does not give optimal scores

    # first we do the rows
    points = 0
    for i in range(6):
        if not np.any(board[i] == 0):
            continue
        for j in range(4):
            row_sum = np.sum(board[i][j:j + 3])
            points += np.sign(row_sum)*row_sum**2

    # then we do the columns
    for i in range(7):
        if not np.any(board[:, i] == 0):
            continue
        for j in range(3):
            column_sum = np.sum(board[j:j + 3, i])
            points += np.sign(column_sum)*column_sum**2

    # then we do the diagonals 
    for i in range(4): # for each column
        for j in range(3): # for each row
            # we take the diagonal from the top left to the bottom right
            # if it is filled with 1s/-1s, we don't add any points

            min_index = min(i, j)
            max_index = max(6 - i, 5 - j)

            if not np.any(np.diag(board[j - min_index:j + max_index + 1, i - min_index:i + max_index + 1])) == 0:
                continue

            diagonal_sum = np.sum(np.diag(board[j:j + 3, i:i + 3]))
            points += np.sign(diagonal_sum)*diagonal_sum**2
            
    for i in range(4): # for each column
        for j in range(3, 6):
            # we take the diagonal from the top left to the bottom right
            # if it is filled with 1s/-1s, we don't add any points

            min_index = min(i, 5 - j)
            max_index = max(6 - i, j)

            if not np.any(np.diag(board[j + min_index:j - max_index - 1:-1, i - min_index:i + max_index + 1])) == 0:
                continue

            diagonal_sum = np.sum(np.diag(board[j - 2:j + 1, i:i + 3]))
            points += np.sign(diagonal_sum)*diagonal_sum**2
    
    
    return points
```
